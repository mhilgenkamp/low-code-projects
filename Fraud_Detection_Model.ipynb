```python
# prompt: import matplotlib.pyplot as plt, numpy as np, pandas as pd, and seaborn as sns

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
url = 'https://storage.googleapis.com/low-code-ai-book/financial_transactions.csv'

transaction_df = pd.read_csv(url)
```

```python
transaction_df.head()
```

```python
transaction_df.info()
```

```python
transaction_df.describe()
```

```python
cols = ['type', 'nameOrig', 'nameDest']
transaction_df[cols].describe()
```

```python
transaction_df['isCorrect'] = (
    transaction_df['isFraud'] ==
transaction_df['isFlaggedFraud']
)
transaction_df['isCorrect'].sum()
```

```python
(transaction_df['isFraud']*transaction_df['isCorrect']).sum()
```

```python
transaction_df.groupby('type')['isFraud'].mean().plot.bar()
```

```python
transaction_df.groupby('type')['isFraud'].value_counts()
```

```python
transaction_df['amountBuckets'] = pd.qcut(transaction_df['amount'], 10)
transaction_df.groupby('amountBuckets')['isFraud'].mean().plot.bar()
```

```python
transaction_df['oldbalanceBuckets'] = pd.qcut(transaction_df['oldbalanceOrg'], 10, duplicates='drop')
transaction_df.groupby('oldbalanceBuckets')['isFraud'].mean().plot.bar()
```
